```python
from sklearn import model_selection,neighbors 
from sklearn.model_selection import train_test_split
A_train,A_test,y_train,y_test=train_test_split(A,y,test_size=0.2,random_state=1)
```

```python
print("Actual test data:")
print(y_test.values)
```

```python
clf=neighbors.KNeighborsClassifier()
clf.fit(A_train,y_train)
```

```python
y_pred=clf.predict(A_test)
```

```python
print("\nPredicted test data:")
print(y_pred)
```

```python
# accuracy score of test data and train data 
from sklearn.metrics import accuracy_score, recall_score, roc_auc_score,confusion_matrix
```

```python
print("\nAccuracy score:%f"%(accuracy_score(y_test,y_pred)*100))
```

```python
print("Recall score:%f"%(recall_score(y_test,y_pred)*100))
```

```python
print("Roc score:%f\n"%(roc_auc_score(y_test,y_pred)*100))
```

```python
print(confusion_matrix(y_test,y_pred))
```
